```python
%matplotlib inline
```

```python
df = pd.read_csv('starbucks.csv',index_col='Date',parse_dates=True)
```

```python
df.head()
```

```python
df['Close'].plot(figsize=(12,5))
```

```python
df.rolling(window=7).mean()
```

```python
df.rolling(window=7).mean().head(10)
```

```python
df['Close'].plot(figsize=(12,5))
df.rolling(window=50).mean()['Close'].plot()
```

```python
df['Close: 30 off'] = df['Close'].rolling(window=30).mean()
```

```python
df.head()
```

```python
df[['Close: 30 off', 'Close']].plot(figsize=(12,5))
```

```python
df['Close'].expanding().mean().plot(figsize=(12,5))
df['Close'].rolling(window=30).mean().plot(figsize=(12,5))
```
